```python
import sys
import codecs
import os
import pandas as pd

def replace_polish_letters_in_file(directory, filename):
    f = codecs.open(directory + "/" + filename, encoding='utf-8')
    contents = f.read()
    new_contents = contents
    f.close()
    
    letters_pl = ["ą", "ę", "ż", "ź", "ń", "ś", "ć", "ł", "ó"]
    letters_eng = ["a", "e", "z", "z", "n", "s", "c", "l", "o"]
    
    new_filename = filename
    for i in range(len(letters_pl)):
        new_contents = new_contents.replace(letters_pl[i], letters_eng[i])
        new_filename = new_filename.replace(letters_pl[i], letters_eng[i])
    
    new_filename = new_filename.replace(" ", "_")
    new_file = open(directory + "/translated/" + new_filename , "w")
    new_file.write(new_contents)
    new_file.close()
    
# import OS module

 
# Get the list of all files and directories
path = "/home/vagrant/projekt/hbase_source"
dir_list = os.listdir(path)
 

for filename in dir_list:
    if filename[-4:] == ".csv":
        replace_polish_letters_in_file(path, filename)
        
        
        
```

```python
 import sys
import codecs
import os

# Get the list of all files and directories
path = "/home/vagrant/projekt/hbase_source"
dir_list = os.listdir(path)
 
    
year_month = [None] * (23*12)
for year in range(23):
    for month in range(12):
        year_month[year*12 + month] = str(year+2000) + "_" + str(month+1) 
    
year_month = pd.Series(year_month, name="Rok_Miesiac")

def transpose_file(path, filename): 
    df = pd.DataFrame(pd.read_csv(path + "/translated/" + filename + "_miesieczne.csv", sep=";")).transpose()
    df.columns = df.iloc[0] + " (" + df.iloc[2] + ")"
    df = df.iloc[3:]
    df.index = year_month[0:df.shape[0]]
    df = df.apply(lambda x: x.str.replace(',','.') )
    df.to_csv(path + "/transposed/" + filename + ".csv", header=1)
    
    
    
chosen_files = ["Rynek_pracy", 
                "Wynagrodzenia_i_swiadczenia_spoleczne", 
                "Handel_zagraniczny"]

for filename in chosen_files:
    transpose_file(path, filename)
        
        
```
